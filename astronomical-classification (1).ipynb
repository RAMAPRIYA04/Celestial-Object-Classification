```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
import keras
```

```python
# Load the dataset
path_to_csv = '/kaggle/input/celestial-object-classification-data/SDSS Tabular Data (1).csv'
df = pd.read_csv(path_to_csv)
```

```python
# Prepare input (X) and output (y) variables
x = df.drop(['objid', 'rerun', 'class'], axis=1)  # Drop unnecessary columns
y = df['class']  # Target column (class: Star, Galaxy, Quasar)
```

```python
# Splitting the datasets into 3 dataframes
star_df = df[df['class']=='STAR']      
qso_df = df[df['class']=='QSO']       
galaxy_df = df[df['class']=='GALAXY']
```

```python
y = y.values.reshape(10000, 1)
```

```python
# Split the data into training and test sets
from re import X
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
# One-hot encode the target variable (y)
enc = OneHotEncoder()
y_train = enc.fit_transform(y_train).toarray()
y_test = enc.transform(y_test).toarray()      
```

```python
# Scale the input features (X) using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
print(X_test.min(), X_test.max())
```

```python
# Build the neural network model
model = Sequential()
model.add(Dense(64, input_dim=15, activation='relu',)) 
model.add(Dense(32, activation='relu'))              
model.add(Dense(16, activation='relu'))              
model.add(Dense(3, activation='softmax'))  
```

```python
# Train the model
history = model.fit(X_train, y_train, validation_data = (X_test, y_test), 
                    epochs=40, batch_size=32)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```
